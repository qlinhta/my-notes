```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
matrix = np.random.normal(0, 1, (100, 10))
```

```python
matrix
```

```python
df = pd.DataFrame(matrix)
```

```python
def rename_colnames(df):
    for i in range(len(df.columns)):
        df.rename(columns={df.columns[i]: 'Feature' + str(i)}, inplace=True)
    return df
```

```python
df = rename_colnames(df)
```

```python
df
```

```python
def heatmap_correlation(df):
    corr = df.corr()
    plt.figure(figsize=(10,10))
    sns.heatmap(corr, annot=True, cmap=plt.cm.Reds)
    plt.show()
    return corr
```

```python
heatmap_correlation(df)
```

```python
def heatmap_correlation_with_mask(df):
    corr = df.corr()
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    plt.figure(figsize=(10,10))
    sns.heatmap(corr, mask=mask, annot=True, cmap=plt.cm.Reds)
    plt.show()
    return corr
```

```python
heatmap_correlation_with_mask(df)
```

```python
def pair_plot(df):
    sns.pairplot(df)
    plt.show()
```

```python
pair_plot(df)
```
