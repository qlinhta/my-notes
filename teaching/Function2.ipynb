```python
from numpy import mean

"""
### Exercise 2:
Write a program that calculates the average score on an exam. Assume we have a small
class of only three students. Assign each student’s score to variables called student1,
student2, and student3 and then use these variables to find the average score. Assign
the average to a variable called average. Print the student scores and the average score.
The following demonstrates the program output when the students have been assigned scores
of 80.0, 90.0, and 66.5:

Student scores:
80.0
90.0
66.5
Average: 78.83333333333333
"""


def get_input():
    student_scores = []
    student1 = float(input("Enter student 1 score: "))
    student2 = float(input("Enter student 2 score: "))
    student3 = float(input("Enter student 3 score: "))
    student_scores.append(student1)
    student_scores.append(student2)
    student_scores.append(student3)
    return student_scores


def average_score(student_scores):
    average = sum(student_scores) / len(student_scores)
    return average


def print_output(student_scores, average):
    print("Student scores:")
    for score in student_scores:
        print(score)
    print("Average: ", average)


student_results = get_input()
average = average_score(student_results)
print_output(student_results, average)
```

```python
"""
### Exercise 3:

Imagine that you teach three classes. These classes have 32, 45, and 51 students. You want to divide the students in these classes into groups with the same number of students in each group but you recognize that there may be some “left over” students.

Assume that you would like there to be 5 groups in the first class (of 32 students), 7 groups in the second class (of
45 students), and 6 groups in the third class (of 51 students). Write a program that uses the
divmod() function to calculate the number of students in each group (where each group
has the same number of students). Print this number for each class and also print the number
of students that will be “leftover” (i.e., the number of students short of a full group). Use
simultaneous assignment to assign the number in each group and the “leftover” to variables.

The following demonstrates the program’s output:

Number of students in each group:
Class 1: 6
Class 2: 6
Class 3: 8
Number of students leftover:
Class 1: 2
Class 2: 3
Class 3: 3
"""


def get_input():
    class1 = int(input("Enter number of students in class 1: "))
    class2 = int(input("Enter number of students in class 2: "))
    class3 = int(input("Enter number of students in class 3: "))
    return class1, class2, class3


def divmod(a, b):
    return a // b, a % b


def print_output(class1, class2, class3, class1_leftover, class2_leftover, class3_leftover):
    print("Number of students in each group:")
    print("Class 1: ", class1)
    print("Class 2: ", class2)
    print("Class 3: ", class3)
    print("Number of students leftover:")
    print("Class 1: ", class1_leftover)
    print("Class 2: ", class2_leftover)
    print("Class 3: ", class3_leftover)


class1, class2, class3 = get_input()
class1, class1_leftover = divmod(class1, 5)
class2, class2_leftover = divmod(class2, 5)
class3, class3_leftover = divmod(class3, 5)
print_output(class1, class2, class3, class1_leftover, class2_leftover, class3_leftover)
```

```python
"""
### Exercise 5:
Write a program that calculates the wavelength of a wave traveling at a constant velocity
given the speed and the frequency. Use the formula λ = v/f, where λ (lambda) is wavelength in meters, v is velocity in meters per second, and f is frequency in Hertz (cycles per second). Print the velocity, frequency, and wavelength. Assign each of these values to a variable and use the variables in your print()statements.
The following demonstrates what the program prints:
The speed (m/s): 343
The frequency (Hz): 256
The wavelength (m): 1.33984375
"""

def get_input():
    speed = float(input("Enter speed (m/s): "))
    frequency = float(input("Enter frequency (Hz): "))
    return speed, frequency


def wavelength(speed, frequency):
    return speed / frequency


def print_output(speed, frequency, wavelength):
    print("The speed (m/s): ", speed)
    print("The frequency (Hz): ", frequency)
    print("The wavelength (m): ", wavelength)


speed, frequency = get_input()
wavelength = wavelength(speed, frequency)
print_output(speed, frequency, wavelength)
```

```python
"""
Exercise 8:
There a list of student names and their grades. Write a program that prints the names of all students who have a grade greater than or equal to 85. The following demonstrates the output from the program:
The students with a grade greater than or equal to 85 are:
    John
    Mary
    Susan
    David
"""

def print_students_with_high_grades(students):
    print("The students with a grade greater than or equal to 85 are:")
    for student in students:
        if student['grade'] >= 85:
            print(student['name'])

students = [{'name': 'John', 'grade': 90}, {'name': 'Mary', 'grade': 85}, {'name': 'Susan', 'grade': 85}, {'name': 'David', 'grade': 90}]
print_students_with_high_grades(students)
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('marketing_campaign.csv', sep=';')
```

```python
df.columns
```

```python
# For all columns name with underscore, replace it with space
for col in df.columns:
    if '_' in col:
        df.rename(columns={col: col.replace('_', ' ')}, inplace=True)
```

```python
# for all columns, convert to lowercase and replace space with underscore
for col in df.columns:
    df.rename(columns={col: col.lower().replace(' ', '_')}, inplace=True)
```

```python
df
```

```python
# Select all people to new dataframe with Education == PhD using for loop with index
new_df = pd.DataFrame()
for i in df.index:
    if df.loc[i, 'education'] == 'PhD':
        new_df = new_df.append(df.loc[i])
new_df
```

```python
# Sort income by descending order
df.sort_values(by='income', ascending=False)
```

```python
# Add a column call IncomeRate, if income > 100,000 then IncomeRate = 1, otherwise 0
df['IncomeRate'] = df['income'].apply(lambda x: 1 if x > 100000 else 0)
```
