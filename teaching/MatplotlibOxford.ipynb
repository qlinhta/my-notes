```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-paper')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font', serif='Times')
plt.rc('axes', titlesize=18)
plt.rc('axes', labelsize=18)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('legend', fontsize=18)
plt.rc('lines', markersize=10)
plt.rc('lines', linewidth=2)
plt.rc('figure', titlesize=18)
plt.rc('figure', figsize=(10, 10))
```

```python
def f(x):
    return np.sin(x)
```

```python
x = np.linspace(0, 3*np.pi, 100)
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, label='$f(x)=\sin(x)$', color='blue', linewidth=1)
# Plot a line at y=0
plt.plot([0, 3*np.pi], [0, 0], color='black', linestyle='--', linewidth=1)
# Plot color zone where y is positive
plt.fill_between(x, y, 0, where=y>0, color='darkorange', alpha=0.2)
# Plot color zone where y is negative
plt.fill_between(x, y, 0, where=y<0, color='darkblue', alpha=0.2)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Sinus of $x$')
plt.legend()
plt.show()
```

```python
# Generate normally distributed data with 100 samples
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.subplots(figsize=(8, 8))
# Scatter plot
plt.scatter(x, y, label='Scatter plot', color='black', marker='+')
# Plot a line at y=0
plt.plot([-3, 3], [0, 0], color='black', linestyle='--', linewidth=2)
plt.fill_between([-3, 3], 3, 0, where=[True, True], color='blue', alpha=0.2)
plt.fill_between([-3, 3], -3, 0, where=[True, True], color='red', alpha=0.2)
plt.xlabel('This is my X')
plt.ylabel('This is my F(X)')
plt.title('Linear function')
plt.legend()
plt.show()
```

```python
# Example of a histogram
x = np.random.normal(0, 1, 1000)
plt.subplots(figsize=(8, 8))
plt.hist(x, bins=20, color='blue', alpha=0.5)
plt.xlabel('This is my X')
plt.ylabel('This is my F(X)')
plt.title('Linear function')
plt.show()
```

```python
# Example of a bar plot
x = ['A', 'B', 'C', 'D', 'E', 'F']
y = [1, -2, 6, 4, 5, -4]
plt.subplots(figsize=(8, 10))
plt.bar(x, y, color=['blue' if i>0 else 'red' for i in y], alpha=0.5)
plt.plot(x, y, color='black', linestyle='--', linewidth=2, marker='D')
plt.xlabel('This is my X')
plt.ylabel('This is my F(X)')
plt.title('Linear function')
plt.show()
```

```python
from math import sqrt

x = [0, 1, 3, 5, 7, 9]
y = [sqrt(i) for i in x]

plt.subplots(figsize=(8, 8))
plt.plot(x, y, color='black', linestyle='solid', linewidth=2, marker='D', label='$f(x)=\sqrt{x}$')
plt.fill_between(x, y, 0, where=[True, True, True, True, True, True], color='grey', alpha=0.2)
plt.vlines(x, 0, y, colors='black', linestyles='dashed', label='Projecting line', linewidth=2, alpha=0.5)
for i in range(len(x)):
    plt.annotate('({}, {})'.format(x[i], round(y[i], 2)), (x[i], y[i]), textcoords='offset points', xytext=(0, 10), ha='center', fontsize=18)
plt.xlabel('$X$')
plt.ylabel('$f(X) = \sqrt{X}$')
plt.title('Linear function $f(x)=\sqrt{x}$')
plt.legend()
plt.show()
```

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D
x = [0, 1, 3, 5, 7, 9]
y = [sqrt(i) for i in x]
z = [i**2 for i in x]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='black', linestyle='solid', linewidth=2, marker='D', label='$f(x)=\sqrt{x}$')
ax.set_xlabel('$X$')
ax.set_ylabel('$f(X) = \sqrt{X}$')
ax.set_zlabel('$f(X)^2 = X$')
ax.set_title('Linear function $f(x)=\sqrt{x}$')
ax.legend()
plt.show()
```

```python
# Scatter plot exercise
x = np.random.normal(0, 1, 500)
y = np.random.normal(0, 1, 500)
plt.subplots(figsize=(8, 8))
plt.scatter(x, y, label='Data', color=['blue' if i>0 else 'red' for i in y], marker='o', alpha=0.5)
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
# Plot distribution of X and Y
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].hist(x, bins=50, color='blue', alpha=0.5)
ax[0].set_xlabel('$X$')
ax[0].set_ylabel('$f(X)$')
ax[0].set_title('Distribution of $X$')
ax[1].hist(y, bins=50, color='blue', alpha=0.5)
ax[1].set_xlabel('$Y$')
ax[1].set_ylabel('$f(Y)$')
ax[1].set_title('Distribution of $Y$')
plt.show()
```
