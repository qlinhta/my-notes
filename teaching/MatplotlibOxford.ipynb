```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-paper')
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('font', serif='Times')
plt.rc('axes', titlesize=18)
plt.rc('axes', labelsize=18)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('legend', fontsize=18)
plt.rc('lines', markersize=10)
plt.rc('lines', linewidth=2)
plt.rc('figure', titlesize=18)
plt.rc('figure', figsize=(10, 10))
```

```python
def f(x):
    return np.sin(x)
```

```python
x = np.linspace(0, 3*np.pi, 100)
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.plot(x, y, label='$f(x)=\sin(x)$', color='black', linewidth=2)
plt.plot([0, 3*np.pi], [0, 0], color='black', linestyle='dashed', linewidth=2)
plt.fill_between(x, y, 0, where=y>0, color='blue', alpha=0.2)
plt.fill_between(x, y, 0, where=y<0, color='red', alpha=0.2)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Sinus of $x$')
plt.legend()
plt.grid()
# Save plot to pdf file
plt.savefig('sinus.png')
plt.show()
```

```python
from math import sqrt

x = [0, 1, 3, 5, 7, 9]
y = [sqrt(i) for i in x]

plt.subplots(figsize=(8, 8))
plt.plot(x, y, color='black', linestyle='solid', linewidth=2, marker='D', label='$f(x)=\sqrt{x}$')
plt.fill_between(x, y, 0, where=[True, True, True, True, True, True], color='grey', alpha=0.2)
plt.vlines(x, 0, y, colors='black', linestyles='dashed', label='Projecting line', linewidth=2, alpha=0.5)
for i in range(len(x)):
    plt.annotate('({}, {})'.format(x[i], round(y[i], 2)), (x[i], y[i]), textcoords='offset points', xytext=(0, 10), ha='center', fontsize=18)
plt.xlabel('$X$')
plt.ylabel('$f(X) = \sqrt{X}$')
plt.title('Linear function $f(x)=\sqrt{x}$')
plt.legend()
plt.show()
```

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D
x = [0, 1, 3, 5, 7, 9]
y = [sqrt(i) for i in x]
z = [i**2 for i in x]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='black', linestyle='solid', linewidth=2, marker='D', label='$f(x)=\sqrt{x}$')
ax.set_xlabel('$X$')
ax.set_ylabel('$f(X) = \sqrt{X}$')
ax.set_zlabel('$f(X)^2 = X$')
ax.set_title('Linear function $f(x)=\sqrt{x}$')
ax.legend()
plt.show()
```

```python
# Plot a gradient descent, animation example
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x = np.linspace(-3, 3, 1000)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y, color='black', linestyle='solid', linewidth=2, label='$f(x)=\sin(x)$')
ax.set_xlabel('$X$')
ax.set_ylabel('$f(X)$')
ax.set_title('Gradient descent')
ax.legend()

line, = ax.plot([], [], color='red', linestyle='solid', linewidth=2, label='$f(x)=\sin(x)$')
point, = ax.plot([], [], color='blue', linestyle='solid', linewidth=2, marker='o', label='$f(x)=\sin(x)$')

def init():
    line.set_data([], [])
    point.set_data([], [])
    return line, point

def animate(i):
    x = np.linspace(-3, 3, 1000)
    y = np.sin(x)
    line.set_data(x, y)
    point.set_data(x[i], y[i])
    return line, point

# Save the animation as a gif
anim = FuncAnimation(fig, animate, init_func=init, frames=1000, interval=20, blit=True)
# MovieWriter imagemagick unavailable; using Pillow instead.
anim.save('gradient_descent.gif', writer='imagemagick', fps=50)

# Show the animation
HTML(anim.to_html5_video())
```
