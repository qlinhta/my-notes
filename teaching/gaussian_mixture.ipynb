```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Make a dataset
X, y = make_blobs(n_samples=1000, n_features=2, centers=1, cluster_std=1.0, random_state=42)
```

```python
# Plot the dataset
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Compute the parameters
mu = np.mean(X, axis=0)
sigma = np.cov(X.T)
pi = X.shape[0] / X.shape[0]
```

```python
# Assign each point to a cluster based on the distance
# from the mean of the cluster
def assign_cluster(X, mu):
    clusters = []
    for i in range(X.shape[0]):
        dist = np.linalg.norm(X[i] - mu)
        if dist < 0.5:
            clusters.append(0)
        else:
            clusters.append(1)
    return clusters

# Plot the clusters
clusters = assign_cluster(X, mu)
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.show()
```
