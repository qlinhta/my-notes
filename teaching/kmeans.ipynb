```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Make a dataset
X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=0)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Initialize the centroids randomly
centroids = np.random.uniform(low=np.min(X), high=np.max(X), size=(3, 2))
```

```python
centroids
```

```python
# Plot the dataset and the centroids
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r')
plt.show()
```

```python
# Calculate the distance between each point and each centroid
distances = np.zeros((X.shape[0], centroids.shape[0]))
for i in range(centroids.shape[0]):
    row_norm = np.linalg.norm(X - centroids[i, :], axis=1)
    distances[:, i] = np.square(row_norm)
```

```python
# Assign all training data to closest (minimum) centroid
clusters = np.argmin(distances, axis=1)
```

```python
# Plot the data and the centroids
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r')
plt.show()
```

```python
# Update the centroids
for i in range(centroids.shape[0]):
    centroids[i, :] = np.mean(X[clusters == i, :], axis=0)

# Plot the data and the centroids
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r')
plt.show()
```

```python
# Repeat the process for 10 iterations
for i in range(10):
    # Calculate the distance between each point and each centroid
    distances = np.zeros((X.shape[0], centroids.shape[0]))
    for i in range(centroids.shape[0]):
        row_norm = np.linalg.norm(X - centroids[i, :], axis=1)
        distances[:, i] = np.square(row_norm)

    # Assign all training data to closest (minimum) centroid
    clusters = np.argmin(distances, axis=1)

    # Update the centroids
    for i in range(centroids.shape[0]):
        centroids[i, :] = np.mean(X[clusters == i, :], axis=0)

# Plot the data and the centroids
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r')
plt.show()
```
