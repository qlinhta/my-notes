```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams.update({'font.size': 14})

dt = 0.001
t = np.arange(0, 2, dt)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2 * np.pi * t * (f0 + (f1 - f0) * np.power(t, 2) / (3 * t1 ** 2)))

plt.specgram(x, NFFT=128, Fs=1 / dt, noverlap=120, cmap='jet')
plt.colorbar()
plt.show()
```
