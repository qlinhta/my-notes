```python

import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray
```

```python
# Import image from datasets folder
img = io.imread('../datasets/Y11.jpg')
```

```python
# Plot photo with matplotlib and seaborn

def plot_photo(img):
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.show()
```

```python
plot_photo(img)
```

```python
# Convert image to grayscale
img_gray = rgb2gray(img)
plot_photo(img_gray)
```

```python
# Zoom in on the image and divide image into 4 parts
img_gray_zoom = img_gray[0:200, 0:200]
plot_photo(img_gray_zoom)
```

```python
x = img.shape[0]
y = img.shape[1]

center_zoomed = img[x // 4:(x // 4 + x // 2), y // 4:(y // 4 + y // 2)]
left_upper = img[:x // 2, :y // 2]
right_upper = img[:x // 2:, y - (y // 2):]
left_lower = img[(x - x // 2):x, :y // 2]
right_lower = img[(x - x // 2):x, y - (y // 2):]

fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=1, ncols=5, figsize=(20, 4))
ax1.imshow(center_zoomed)
ax2.imshow(left_upper)
ax3.imshow(right_upper)
ax4.imshow(left_lower)
ax5.imshow(right_lower)
plt.show()
```

```python
# Display RGB layers of image

def plot_RGB(img):
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 4))
    ax1.imshow(img[:, :, 0])
    ax2.imshow(img[:, :, 1])
    ax3.imshow(img[:, :, 2])
    plt.show()
```

```python
plot_RGB(img)
```

```python
# Invert colr of images

def invert_color(img):
    return 255 - img
```

```python
inverted_img = invert_color(img)
```

```python
plot_photo(inverted_img)
```

```python
# Color segmentation of images

def color_segmentation(img):
    # Create a mask for each color
    red_mask = img[:, :, 0] > 200
    green_mask = img[:, :, 1] > 200
    blue_mask = img[:, :, 2] > 200
    # Combine masks
    mask = red_mask & green_mask & blue_mask
    # Apply mask to image
    return img[mask]
```

```python
color_segmentation(img)
```

```python
# Create a function to create a mask for each color
def create_mask(img, color):
    if color == 'red':
        return img[:, :, 0] > 200
    elif color == 'green':
        return img[:, :, 1] > 200
    elif color == 'blue':
        return img[:, :, 2] > 200
    else:
        return None
```
